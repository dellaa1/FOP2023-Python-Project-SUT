```python
def ln71():
    sorted_df = df.sort_values(by='Area')
    sorted_df['Area'] = pd.to_numeric(sorted_df['Area'], errors='coerce')
    sorted_df['Price(USD)'] = pd.to_numeric(sorted_df['Price(USD)'], errors='coerce')
    sorted_df.dropna(subset=['Area', 'Price(USD)'], inplace=True)
    plt.figure(figsize=(10, 6))
    plt.plot(sorted_df['Area'], sorted_df['Price(USD)'], color='green', linestyle='--', marker='o',
             label='Price(USD) by Area')
    coefficients = np.polyfit(sorted_df['Area'], sorted_df['Price(USD)'], 1)
    trend_line = np.poly1d(coefficients)
    plt.plot(sorted_df['Area'], trend_line(sorted_df['Area']), color='blue', label='Linear Trend Line')
    linear_changes_line = coefficients[0] * sorted_df['Area'] + coefficients[1]
    plt.plot(sorted_df['Area'], linear_changes_line, color='red', linestyle='--',
             label='Linear Changes (Red Dashed Line)')
    plt.xlabel('Area')
    plt.ylabel('Price(USD)')
    plt.title('Broken Line Chart with Linear Trend Line and Linear Changes')
    plt.legend()
    plt.show()
ln71()
```
